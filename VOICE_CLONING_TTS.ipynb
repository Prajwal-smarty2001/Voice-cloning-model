```python

# import all libraries

import os, sys
import subprocess
import tempfile
from time import time
import warnings
from cog import BasePredictor, Path
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F
import IPython
from IPython.display import clear_output
from tortoise.api import TextToSpeech, format_conditioning
from tortoise.utils.audio import load_audio, get_voices, load_voice, load_voices
from tortoise.utils.tokenizer import VoiceBpeTokenizer
from tortoise.utils.text import split_and_recombine_text
from tortoise.api import TextToSpeech
```

```python

# set the voice paths
CUSTOM_VOICE_FILE = Path(CUSTOM_VOICE_DIRECTORY, file_name)
outpath = "/content/results/"
textfile_path = "../custom_speech.txt"
CUSTOM_VOICE_NAME = voice_to_use
voice_outpath = os.path.join(outpath, CUSTOM_VOICE_NAME)
```

```python
# Function to create custom_voice from .mp3 file
def create_custom_voice_from_mp3(input_path: str, segment_time: int = 9) -> None:
    CUSTOM_VOICE_DIRECTORY.mkdir(parents=True, exist_ok=True)
    normalized_audio_path = tempfile.mktemp(suffix=".wav")
    result = subprocess.run([
            "ffmpeg-normalize",
            input_path,
            "-c:a",
            "libmp3lame",
            "-b:a",
            "192k",
            "-o",
            normalized_audio_path,], stdout=subprocess.PIPE)
    try:
      output = subprocess.check_output(
        [
            "ffmpeg-normalize",
            input_path,
            "-c:a",
            "libmp3lame",
            "-b:a",
            "192k",
            "-o",
            normalized_audio_path,
        ],
        stderr=subprocess.STDOUT,
    )
    except subprocess.CalledProcessError as e:
      print(e.output.decode('utf-8'))
    assert os.path.exists(normalized_audio_path), "ffmpeg-normalize failed"
    result = subprocess.run(   [
            "ffmpeg",
            "-v",
            "warning",  # log only errors
            "-ss",
            "00:00:00",
            "-t",
            #"300", # limit to 300 seconds (~30 segments)
            "500", # limit to 300 seconds (~40 segments)
            #"500",  # limit to 300 seconds (~50 segments)
            "-i",
            normalized_audio_path,
            "-acodec",
            "pcm_s16le",
            "-ac",
            "1",
            "-f",
            "segment",
            "-segment_time",
            str(segment_time),
            os.path.join(CUSTOM_VOICE_DIRECTORY, "%d.wav"),
        ])

    subprocess.check_output(
        [
            "ffmpeg",
            "-v",
            "warning",  # log only errors
            "-ss",
            "00:00:00",
            "-t",
            #"300", # limit to 300 seconds (~30 segments)
            "500",  # limit to 300 seconds (~30 segments)
            "-i",
            normalized_audio_path,
            "-acodec",
            "pcm_s16le",
            "-ac",
            "1",
            "-f",
            "segment",
            "-segment_time",
            str(segment_time),
            os.path.join(CUSTOM_VOICE_DIRECTORY, "%d.wav"),
        ]
    )
```

```python
# init of the text variable
text = ""
```

```python


class Predictor(BasePredictor):
    def predict(
        text_to_say: str = text,
        voice_a: str = voice_to_use,
        custom_voice: Path = CUSTOM_VOICE_FILE,
        voice_b: str = "disabled",
        voice_c: str = "disabled",
        preset_to_use: str = preset_to_use
            #["ultra_fast", "fast", "standard", "high_quality"],
        ,
        seed: int= (time()),
        cvvp_amount: float = cvvp_amount_to_use
        ) -> Path:

        output_dir = Path(tempfile.mkdtemp())

        if custom_voice is not None:
            assert (
                custom_voice.suffix == ".mp3"
            ), f"File {custom_voice} is not an mp3 file"
            if voice_to_use == "custom_voice":
              create_custom_voice_from_mp3(str(custom_voice))
              print(f"Creating voice from {custom_voice}")
            else:
              pass
            all_voices = [voice_to_use]
        else:
            all_voices = [voice_to_use]
        if voice_b != "disabled":
            all_voices.append("custom_voice")
        if voice_c != "disabled":
            all_voices.append(voice_c)
        print(f"Generating text using voices: {all_voices}")




        # Textfile process
        with open(textfile_path, 'r', encoding='utf-8') as f:
            text = ' '.join([l for l in f.readlines()])
            if '|' in text:
                print("Found the '|' character in your text, which I will use as a cue for where to split it up. If this was not"
                      "your intent, please remove all '|' characters from the input.")
                texts = text.split('|')
            else:
                texts = split_and_recombine_text(text)

        seed = 0 #int(time())

        os.makedirs(voice_outpath, exist_ok=True)

        voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
        all_parts = []
        # generating your custom_voice reading
        for j, text in enumerate(texts):
            gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                                      preset=preset_to_use, k=1, use_deterministic_seed=seed,cvvp_amount=cvvp_amount)
            gen = gen.squeeze(0).cpu()
            torchaudio.save(os.path.join(voice_outpath, f'{j}.wav'), gen, 24000)
            all_parts.append(gen)

        # save your custom_voice reading to "combined.wav" file
        full_audio = torch.cat(all_parts, dim=-1)
        torchaudio.save(os.path.join(voice_outpath, 'result3.wav'), full_audio, 24000)

def exit(code):
    #print('\nExiting....')
    #print('\nHave a good day.')
    sys.exit(code)

def main():
    try:
      screen_clear()
      Predictor.predict()
      screen_clear()
      #print("\nDone !  Your file is saved on your Google Drive !")
    except KeyboardInterrupt:
      exit(1)





```
